```python
import os
import cv2
import numpy as np

for i in ["train", "valid", "test"]:
    print(i)
    # Define the paths
    source_images_dir = f"C:\\Users\\Lenovo\\Desktop\\test\\datasets\\Adobe-GameSense-5\\{i}\\images"
    dest_images_dir = f"C:\\Users\\Lenovo\\Desktop\\test\\datasets\\Adobe-GameSense-5\\{i}\\thresholded_images"

    # Create destination directory if it doesn't exist
    if not os.path.exists(dest_images_dir):
        os.makedirs(dest_images_dir)

    # Process each image in the source directory
    for filename in os.listdir(source_images_dir):
        image_path = os.path.join(source_images_dir, filename)
        
        # Check if the file is an image
        if not os.path.isfile(image_path):
            continue

        # Read the image
        image = cv2.imread(image_path)
        if image is None:
            print(f"Error: Image {filename} not loaded properly.")
            continue

        # Convert the image to grayscale to apply intensity-based thresholding
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Create a mask: if pixel intensity < 180, set the mask to 0 (black); otherwise, set to 255 (white)
        _, mask = cv2.threshold(gray_image, 100, 255, cv2.THRESH_BINARY)

        # Apply the mask: where the mask is 0, make the original image black, otherwise leave it unchanged
        thresholded_image = cv2.bitwise_and(image, image, mask=mask)

        # Save the annotated image
        dest_image_path = os.path.join(dest_images_dir, filename)
        cv2.imwrite(dest_image_path, thresholded_image)

    print(f"Processed images saved in {dest_images_dir}")
```

```python
import os
import cv2

def draw_bounding_box(image, x_center, y_center, width, height, color=(0, 255, 0), thickness=2):
    """
    Draws a bounding box on the image based on the YOLO format.
    """
    h, w, _ = image.shape
    x_center_pixel = int(x_center * w)
    y_center_pixel = int(y_center * h)
    width_pixel = int(width * w)
    height_pixel = int(height * h)

    top_left = (x_center_pixel - width_pixel // 2, y_center_pixel - height_pixel // 2)
    bottom_right = (x_center_pixel + width_pixel // 2, y_center_pixel + height_pixel // 2)

    cv2.rectangle(image, top_left, bottom_right, color, thickness)

for i in ["train", "valid", "test"]:
    print(f"Processing {i} dataset")

    # Define the paths
    source_images_dir = f"/kaggle/working/datasets/Adobe-GameSense-1/{i}/preprocessed_images"
    source_labels_dir = f"/kaggle/working/datasets/Adobe-GameSense-1/{i}/preprocessed_labels"
    dest_images_dir = f"/kaggle/working/datasets/Adobe-GameSense-1/{i}/images_with_boxes"

    # Create destination directory if it doesn't exist
    if not os.path.exists(dest_images_dir):
        os.makedirs(dest_images_dir)

    # Process only the first image in the source directory
    for idx, filename in enumerate(os.listdir(source_images_dir)):
        if idx > 5:  # Process only the first image
            break

        image_path = os.path.join(source_images_dir, filename)
        label_path = os.path.join(source_labels_dir, os.path.splitext(filename)[0] + '.txt')

        if not os.path.isfile(image_path) or not os.path.isfile(label_path):
            continue

        # Read the image
        image = cv2.imread(image_path)
        if image is None:
            print(f"Error: Image {filename} not loaded properly.")
            continue

        # Read the annotation file and draw bounding boxes
        with open(label_path, 'r') as f:
            lines = f.readlines()

        for line in lines:
            parts = line.strip().split()
            class_id = parts[0]
            x_center = float(parts[1])
            y_center = float(parts[2])
            width = float(parts[3])
            height = float(parts[4])

            # Draw bounding box on the image
            draw_bounding_box(image, x_center, y_center, width, height)

        # Save the image with bounding boxes
        dest_image_path = os.path.join(dest_images_dir, filename)
        cv2.imwrite(dest_image_path, image)

        print(f"Saved {filename} with bounding boxes in {i} dataset.")
```

```python
import cv2
import pandas as pd
from ultralytics import YOLO

# Load the model
model = YOLO('/kaggle/working/runs/detect/train6/weights/best.pt')

# Open the video file
cap = cv2.VideoCapture('/kaggle/working/video/test/12sec.mp4')

# Ensure video opened successfully
if not cap.isOpened():
    print("Error opening video file")
    exit()

# Load class names (replace with actual class names for your model)
class_list = ['shuttle']  # Update this with your actual class names

# Resize dimensions to match model input size if necessary
resize_width, resize_height = 640, 640  # Example resize dimensions; update as needed

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Resize frame to match model input size
    frame_resized = cv2.resize(frame, (resize_width, resize_height))
    
    # Perform object detection
    results = model.predict(frame_resized)
    a = results[0].boxes.data
    a = a.detach().cpu().numpy()
    
    # Convert detections to DataFrame
    px = pd.DataFrame(a, columns=['x1', 'y1', 'x2', 'y2', 'confidence', 'class_id']).astype("float")

    # Convert bounding box coordinates from resized image to original size
    original_height, original_width = frame.shape[:2]
    x_scale = original_width / resize_width
    y_scale = original_height / resize_height

    # Loop through detections and draw bounding boxes
    for index, row in px.iterrows():
        x1, y1, x2, y2, confidence, class_id = row
        class_id = int(class_id)
        class_name = class_list[class_id]
        
        if 'shuttle' in class_name:
            # Convert bounding box coordinates to original frame size
            x1 = int(x1 * x_scale)
            y1 = int(y1 * y_scale)
            x2 = int(x2 * x_scale)
            y2 = int(y2 * y_scale)
            
            # Draw bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            # Put class label
            label = f'{class_name} {confidence:.2f}'
            cv2.putText(frame, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    # Display the frame with bounding boxes
    cv2.imshow('Detection', frame)
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
